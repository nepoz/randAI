```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from pprint import pprint
```

```python
## Read in iris data from csv, display tail to make sure we're on the right track

iris = pd.read_csv('iris.data', header=None)
iris.tail()
```

```python
## Getting values and getting a scatter plot for sepal width vs. petal length

setosas = iris.iloc[0:50]
versicolors = iris.iloc[50:100]
virginicas = iris.iloc[100:]

plt.scatter(setosas[1], setosas[2], color='r', label="Iris-setosa")
plt.scatter(versicolors[1], versicolors[2], color='b', label="Iris-versicolor")
plt.scatter(virginicas[1], virginicas[2], color='g', label="Iris-virginica")
plt.xlabel("Sepal width in cm")
plt.ylabel("Petal length in cm")
plt.legend(loc='right')

plt.show()
```

```python
'''It seems that on the basis of features sepal width and petal length, iris setosa is linearly separable
For simplicity, we will be looking at flowers that are either setosa or versicolor.'''

## Features that will be looked at: Sepal width, petal length
setosa_features = list(zip(setosas[1].values, setosas[3].values))
versicolor_features = list((zip(versicolors[1].values, versicolors[3].values)))
input_features = np.array(setosa_features + versicolor_features)

## Cleaning up versicolor and setosa results to be either 1 or -1
setosa_outputs = np.where(setosas[4].values, 1, -1)
versicolor_outputs = np.where(versicolors[4].values, -1, 1)
ground_truths = np.append(setosa_outputs, versicolor_outputs)
```

```python
'''Here, we provide the Perceptron class that will be our model'''
class Perceptron(object):
    
    def __init__(self, learning_rate=0.01, random_seed=1, max_iter=50):
        self.learning_rate = learning_rate
        self.random_seed = random_seed,
        self.max_iter = max_iter
        
    ## Method that fits model to specific training data
    def fit(self, features, ground_truths):
        train = list(zip(features[0:70], ground_truths[0:70]))
        test = np.array(list(zip(features[30:], ground_truths[30:])))
        np.random.seed(self.random_seed)

        self.w_ = np.random.rand(len(features[0]) + 1)
        epochs = 0
        while (epochs < self.max_iter):
            errors = 0
            for xi, y in train:
                output = self.predict(xi)
                error = output - y
                
                ## Update the weights
                for i in range(1, len(xi) + 1):
                    self.w_[i] += self.learning_rate * error * xi[i - 1]
                self.w_[0] += self.learning_rate * error
                
            for xi, y in test:
                output = self.predict(xi)
                error = output - y
                
                if error != 0:
                    errors += 1
                    
            epochs += 1
        
        return self
        
        
    def net_sum(self, weights, features):
        return (np.dot(weights[1:], features) + weights[0])
    
    def predict(self, features):
        net_sum = self.net_sum(self.w_, features)
        
        if net_sum > 0:
            return 1
        else:
            return - 1
```

```python
'''The big test, let us see if our model can actually do something'''
ppn = Perceptron()
ppn = ppn.fit(input_features, ground_truths)
```

```python
ppn.predict((2.2, 1.5))
```
