```python
import cv2
from matplotlib import pyplot as plt
```

```python
## Read in a test image in color
test_img = cv2.imread('test.png', cv2.IMREAD_COLOR)
```

```python
## Convert to LAB colorspace
LAB = cv2.cvtColor(test_img, cv2.COLOR_BGR2LAB)
```

```python
## Split channels, and visualize what each looks like by itself

L, A, B = cv2.split(LAB)

plt.subplot(2, 2, 1)
plt.title('Original')
plt.axis('off')
plt.imshow(cv2.cvtColor(LAB, cv2.COLOR_LAB2RGB))

plt.subplot(2, 2, 2)
plt.title('Lightness (input)')
plt.axis('off')
plt.imshow(L, cmap='gray')

plt.subplot(2, 2, 3)
plt.title('A* (one target)')
plt.axis('off')
plt.imshow(A, cmap='gray')

plt.subplot(2, 2, 4)
plt.title('B* (other target)')
plt.axis('off')
plt.imshow(B, cmap='gray')

plt.rcParams["figure.dpi"] = 150
plt.tight_layout()
plt.show()
```
